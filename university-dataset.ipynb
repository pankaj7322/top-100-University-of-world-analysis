```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, dpi = 120, figsize = (8,4))
plot00 = sns.distplot(dataset['overall_score'], ax= axes[0], color ='#E29F7A' )
axes[0].set_title('Overall Score', fontdict = {'fontsize':8})
axes[0].set_xlabel('Overall Score', fontdict = {'fontsize':7})
axes[0].set_ylabel('Frequency/Distribution', fontdict = {'fontsize':7})
plt.tight_layout()
 
plot01 = sns.boxplot(data = dataset, x= 'overall_score', ax=axes[1], orient='h', color ='#E29F7A')
axes[1]. set_title('overall_score', fontdict = {'fontsize':8})
plt.tight_layout()


```

```python
# find mean
mean = str(dataset['overall_score'].mean())
print("The mean score for overall_score is:" + mean)

# find median
median = str(dataset['overall_score'].median())
print("The median score for overall_score is: " + median)
```

- A mean score would indicate it is highly scoring dataset, which would make a lot of sense given its the top 100 universities in the world
- Given the median is slightly below the mean, especially given we know there are some high scoring universities in the dataset


```python
# find the skewness of the dataset

print("Skewness: %f" % dataset['overall_score'].skew())
print("Kurtosis: %f" % dataset['overall_score'].kurt())
```

- The skewness score of positive but close to 0 would indicate that there are some exceptional socre in the set but the distribution is largely normally distributed. This would be expected given the top 5 in this dataset are significantly better than the rest as already highlighted
- The negative kurtosis would indicate that there are lighter tails in the set which is theory means the overall shape is flatter

```python
fig,axes1 = plt.subplots(nrows = 1, ncols=2, dpi=120, figsize = (8,4))
plot11 = sns.displot(dataset['sustainability'], ax= axes1[0], color='#E29F7A')
axes1[0].set_title('Sustainability',fontdict={'fontsize':8})
axes1[0].set_xlabel('Sustainability',fontdict={'fontsize':7})
axes1[0].set_ylabel('Frequency/Distrubtion',fontdict={'fontsize':7})
plt.tight_layout()

plot02=sns.boxplot(data=dataset, x='sustainability',ax=axes[1],orient = 'h',color='#E29F7A')
axes[1].set_title('sustainability',fontdict={'fontsize':8})
plt.tight_layout()


```

```python
mean = str(dataset["sustainability"].mean())
print("The mean score for sustainability is: " + mean) 

median = str(dataset["sustainability"].median())
print("The median score for sustainability is: " + median)
```

 - Again the score for sustainability is high, with an average of 78. I'd expect this, with eyes being on the universitiy, as well as (most likely) having the highest amount of funding/money to be able to focus on this
 - The median is much higher than the mean, given this mean half of the values are above the score of 83.7, it would suggest the data is positively skewed. This is easy to see in the charts above, but we should check the scores anyway
